```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import scipy.fftpack as spfft
import scipy.signal as spsig
from scipy.io import wavfile
```

```python
wav_path = './SampleSounds/401824__logicmoon__dubbish-108.wav'
wav_file = wavfile.read(wav_path)
wf_bd_max = np.iinfo(np.int16).max
wf_fs = wav_file[0]
wf_x1 = wav_file[1][:, 0].flatten()
wf_x2 = wav_file[1][:, 1].flatten()
print(wf_fs)
```

```python
fig, axs = plt.subplots(2)
fig.set_figheight(8)
fig.set_figwidth(15)
axs[0].plot(wf_x1)
axs[1].plot(wf_x2)
plt.show()
```

```python
window = 'blackmanharris'
freq_vals, time_vals, spg = spsig.stft(wf_x1, fs=wf_fs, window=window, nperseg=1024)
```

```python
plt.figure(figsize=(10, 6))
plt.pcolormesh(time_vals, freq_vals, 2 * np.log10(np.abs(spg)), cmap='inferno')
```

```python
plt.figure(figsize=(6, 6))
plt.pcolormesh(time_vals[:100], freq_vals[:100], 2 * np.log10(np.abs(spg[:100, :100])), cmap='inferno')
```

```python
print(time_vals.shape)
```

```python
spg_segment = spg[1:, :512]
```

```python
pic_full_fft = np.zeros((1024, 1024), dtype=np.complex64)
pic_full_fft[:512, :512] = spg_segment
pic_full_fft[512:, :512] = np.rot90(spg_segment, 1)
pic_full_fft[512:, 512:] = np.rot90(spg_segment, 2)
pic_full_fft[:512, 512:] = np.rot90(spg_segment, 3)
plt.figure(figsize=(6, 6))
plt.imshow(np.log10(np.abs(pic_full_fft)), cmap='inferno')
plt.show()
```

```python
pic_full = np.fft.ifft2(pic_full_fft[:512, :512])
pic_full_abs = np.abs(pic_full)
pic_full_abs /= pic_full_abs.max()

plt.figure(figsize=(6, 6))
plt.imshow(np.log2(pic_full_abs + 1) * 255, cmap='inferno')
```

```python
pic_full_rot = np.rot90(pic_full, 1)
re_times, re_vals = spsig.istft(pic_full_rot, fs=wf_fs, window=window)
```

```python
re_signal_casted = np.float32(re_vals) * 1000
wavfile.write('./1.wav', wf_fs, re_signal_casted)
```
